```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('result/words-embeddings.csv')
df['embedding'] = df['embedding'].apply(eval).apply(np.array)
df
```

```python
from openai.embeddings_utils import get_embedding, cosine_similarity
```

```python
search_term = 'red'
search_term_vector = get_embedding(search_term, engine='text-embedding-ada-002')
print(search_term_vector)

df['similarities'] = df['embedding'].apply(lambda x: cosine_similarity(x, search_term_vector))
df = df.sort_values('similarities', ascending=False)
df
```

### Other example

```python
df = pd.read_csv('result/words-embeddings.csv')
df['embedding'] = df['embedding'].apply(eval).apply(np.array)

v1 = df['embedding'].iloc[10]   # cappuccino
v2 = df['embedding'].iloc[21]   # milk

v = v2 + v1
# v
df['similarities'] = df['embedding'].apply(lambda x: cosine_similarity(x, v))
df.sort_values('similarities', ascending=False)
df
```
